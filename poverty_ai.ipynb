```python
import pandas as pd

data = {
    "income": [5000, 15000, 3000, 25000, 7000, 20000],
    "family_members": [6, 4, 7, 3, 5, 4],
    "employed": [0, 1, 0, 1, 0, 1],
    "poverty": [1, 0, 1, 0, 1, 0]
}

df = pd.DataFrame(data)
df
```

```python
from sklearn.tree import DecisionTreeClassifier

X = df[['income', 'family_members', 'employed']]
y = df['poverty']

model = DecisionTreeClassifier()
model.fit(X, y)

print("Model trained successfully!")
```

```python
income = 8000
family_members = 5
employed = 0

prediction = model.predict([[income, family_members, employed]])

if prediction[0] == 1:
    print("High Poverty Risk")
else:
    print("Low Poverty Risk")
```

```python
income = int(input("Enter monthly income: "))
family_members = int(input("Enter family members: "))
employed = int(input("Employed? (1=Yes, 0=No): "))

prediction = model.predict([[income, family_members, employed]])

if prediction[0] == 1:
    print("Prediction: High Poverty Risk")
else:
    print("Prediction: Low Poverty Risk")
```

```python
from sklearn.metrics import accuracy_score

predictions = model.predict(X)
accuracy = accuracy_score(y, predictions)

print("Accuracy:", accuracy * 100, "%")
```
